```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read filecsv
netflix = pd.read_csv("Netflix Userbase.csv")
#set index
netflix.set_index('User ID' , inplace = True)
print(netflix)
print("----------------------------------------------------")
print(netflix.shape)
print("----------------------------------------------------")
print(netflix.info())
```

```python
#create a pandas groupby data frame
countries = netflix.groupby(['Country'])
print(type(countries))
```

```python
#เรียกดูยอดสรุปข้อมูล
viewsum = ['Gender', 'Subscription Type' , 'Country' , 'Device']
for col in viewsum :
    print(col, netflix[col].value_counts().sort_index(), sep="\n\n", end = "\n\n")
```

```python
# review some descriptive statistics
print(netflix.groupby('Gender').describe())
```

```python
cc = 'Australia','Brazil','Canada','France','Germany','Italy','Mexico','Spain','United Kingdom','United States'
dd = [183,183,317,183,183,183,183,451,183,451]

plt.bar(cc,dd, color = 'g')

plt.title("Count of member Netflix in the world")
plt.xlabel("cc")
plt.ylabel("dd")
plt.rcParams["figure.figsize"] = (20, 10)
plt.show()
```

```python
#
import matplotlib.pyplot as plt
labels = 'Basic', 'Premium' , 'Standard'
sections = [999 , 733, 768]

colors =['yellow','chocolate', 'deeppink']
plt.pie(sections, labels = labels, colors=colors,
       startangle = 90,
       explode = (0.1,0.1,0.1))

plt.legend(sections)
plt.axis('equal')
plt.title("Subscription Type Popular")
plt.show()
```
